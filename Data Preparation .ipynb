```python
import pandas as pd 
import re
df = pd.read_csv("./data/RGeo_Hotel_Reviews.csv", index_col=0)
```

```python
#review negative and postitive reviews in separate dataframe
df_negative_review = df.Negative_Review
df_positive_review = df.Positive_Review
```

```python
df_negative_review
```

```python
#change all negative reviews into lowercase and only words are kept
neg_review_lower = [[word.lower() for word in neg_review.split()] for neg_review in df_negative_review]
neg_review_alpha = [[word for word in sentence if re.search('^[a-z]+$', word)] for sentence in neg_review_lower]
```

```python
#change all positive reviews into lowercase and only words are kept
pos_review_lower = [[word.lower() for word in pos_review.split()] for pos_review in df_positive_review]
pos_review_alpha = [[word for word in sentence if re.search('^[a-z]+$', word)] for sentence in pos_review_lower]
```

```python
#merging updated words back to a sentence
df['cleaned_pos_reviews'] = [' '.join(sentence) for sentence in pos_review_alpha]
df['cleaned_neg_reviews'] = [' '.join(sentence) for sentence in neg_review_alpha]
```

# Replace unwanted comments into '0'

```python
# replace all the ['no negative', 'no positive','n a','na','no issue','no complaints','none','nothing'] with NaN
cleaned_df=df.replace(to_replace=['no negative', 'no positive',
                                  'n a','na','no issue','no complaints',
                                  'none','nothing',''], value=0)
```

```python
cleaned_df
```

# Pull only positive reviews and add Class (positive)

```python
pos_comments= cleaned_df.cleaned_pos_reviews[cleaned_df.cleaned_pos_reviews!=0]
reviews=pd.DataFrame(pos_comments)
reviews['class']='positive'
reviews.columns = ['reviews','class']
reviews
```

# Pull only negative reviews and add Class (negative)

```python
neg_comments= cleaned_df.cleaned_neg_reviews[cleaned_df.cleaned_neg_reviews!=0]
neg_reviews= pd.DataFrame(neg_comments)
neg_reviews['class']= 'negative'
neg_reviews.columns = ['reviews','class']
neg_reviews
```

# Merge both negative and positive and create output file

```python
merge_pos_neg= pd.concat([neg_reviews, reviews])
```

```python
merge_pos_neg.to_csv('./data/cleaned_hotelreviews.csv', index=False)
```
