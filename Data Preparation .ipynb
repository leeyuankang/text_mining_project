```python
import pandas as pd 
import re
df = pd.read_csv("./data/RGeo_Hotel_Reviews.csv", index_col=0)
```

```python
df_negative_review = df.Negative_Review
df_positive_review = df.Positive_Review
```

```python
df_negative_review
```

```python
#change all negative reviews into lowercase and only words are kept
neg_review_lower = [[word.lower() for word in neg_review.split()] for neg_review in df_negative_review]
neg_review_alpha = [[word for word in sentence if re.search('^[a-z]+$', word)] for sentence in neg_review_lower]
```

```python
#change all positive reviews into lowercase and only words are kept
pos_review_lower = [[word.lower() for word in pos_review.split()] for pos_review in df_positive_review]
pos_review_alpha = [[word for word in sentence if re.search('^[a-z]+$', word)] for sentence in pos_review_lower]
```

```python
#merging updated words back to a sentence
df['cleaned_pos_reviews'] = [' '.join(sentence) for sentence in pos_review_alpha]
df['cleaned_neg_reviews'] = [' '.join(sentence) for sentence in neg_review_alpha]
```

```python
#convert positive and negative reviews from csv file to text file
with open('./data/cleaned_hotel_reviews.csv', 'r') as file:
    data = file.read().split('\n')
data
```

```python
len(data)
```

```python
for row in range(1, len(data)):
    reviews= data[row].split(',')

    #create positive text file if it is not 'NaN'
    directory= './hotelreviews_text_files/'
    if reviews[-2]!='NaN':
        with open(directory + 'pos' + str(row) + '.txt', 'w') as output:
            output.write(reviews[-2])
    
    #create negative text file if it is not 'NaN'
    if reviews[-1]!='NaN':
        with open(directory + 'neg' + str(row) + '.txt', 'w') as output:
            output.write(reviews[-1])
```
