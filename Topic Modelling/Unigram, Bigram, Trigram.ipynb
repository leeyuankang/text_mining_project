```python
import pandas as pd
import gensim
```

```python
# Read data
df=pd.read_csv('./data/cleaned_hotelreviews.csv')
df=df.dropna()
df['reviews'] = df['reviews'].str.split(" ")
```

```python
doc = df['reviews'].values.tolist()
doc
```

# BIGRAM

```python
bigram = gensim.models.phrases.Phrases(doc, min_count=3, threshold=10)
print(bigram[doc[0]])
```

# TRIGRAM

```python
trigram = gensim.models.phrases.Phrases(bigram[doc], threshold=10)
print(trigram[bigram[doc[0]]])
```

# QUADGRAM

```python
quadgram = gensim.models.phrases.Phrases(trigram[doc], threshold=10)
print(quadgram[trigram[doc[0]]])
# eg. trigram will return all the unigram, bigram and trigram
```
